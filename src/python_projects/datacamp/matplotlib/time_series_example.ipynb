```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```

```python
sixties = climate_change["1960-01-01":"1969-12-31"]
```

```python
fig, ax = plt.subplots()
ax.plot(sixties.index, sixties['co2'])
ax.set_xlabel('Year in 1960s')
ax.set_ylabel('CO2 (ppm)')
x_markers = ["1960-01-06","1961-01-06", "1962-01-06", "1963-01-06", "1964-01-06", "1965-01-06", "1966-01-06", "1967-01-06","1968-01-06","1969-01-06","1970-01-06"]
x_labels = ["1960", "1961", "1962", "1963", "1964", "1965", "1966", "1967", "1968", "1969", "1970"]
plt.xticks(x_markers, x_labels)
plt.show()
```

```python
sixty_nine = climate_change["1969-01-01":"1969-12-31"]
fig, ax = plt.subplots()
ax.plot(sixty_nine.index, sixty_nine['co2'])
ax.set_xlabel('CO2 (ppm) in the month of 1969')
x_markers = ["1969-01-06","1969-02-06", "1969-03-06", "1969-04-06", "1969-05-06", "1969-06-06", "1969-07-06", "1969-08-06","1969-09-06","1969-10-06","1969-11-06", "1969-12-06"]
x_labels = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
plt.xticks(x_markers, x_labels, rotation=45)
plt.show()
```
