```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
df = pd.read_csv("countries-of-the-world.csv")
```

```python
df.head()
```

```python
gdp = df["GDP ($ per capita)"]
```

```python
phones = df["Phones (per 1000)"]
```

```python
percent_literate = df["Literacy (%)"]
```

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp, y=phones)
plt.show()
```

```python
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=percent_literate)
plt.show()
```

```python
# replace missing values with 0.0
percent_literate = percent_literate.fillna(0.0)
percent_literate = percent_literate.str.replace(",", ".", regex=False)
percent_literate = percent_literate.astype(float)
```

```python
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=percent_literate)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Example using scatterplot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
# Using relplot()
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="smoker")
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", row="smoker")
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="smoker", row="time")
plt.show()
```

```python
col_order = ["Sun", "Sat", "Thur", "Fri"]
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="day", col_order=col_order)
plt.show()
```

```python
col_order = ["Sun", "Sat", "Thur", "Fri"]
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="day", col_wrap=2, col_order=col_order)
plt.show()
```
