### Exercise: "Planning and Creating a Pivot Table"

Create a pivot table that shows the highest sale price for each sales region by product category. This time, each sales
region should be in a separate columnm, and each product category in a separate row. For this example you can ignore
the overall total rows and columns, but you should appy a currency format using the $ symbol and use the underscore (_) character as your thousands separator. Again, two decimal places are needed.

```python
pd.set_option('display.float_format', '${:_.2f}'.format)
sales_data.pivot_table(values="sale_price", index="product_category", columns="sales_region", aggfunc="max")
```

```python
pd.set_option('display.float_format', '${:,.2f}'.format)
sales_data.pivot_table(values='sale_price', index='customer_state', columns=["customer_type", "order_type"], fill_value=0)
```
