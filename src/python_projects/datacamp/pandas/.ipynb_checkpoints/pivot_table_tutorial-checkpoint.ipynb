```python
import pandas as pd
sales_data = pd.read_csv('sales_data2.csv', dayfirst=True).convert_dtypes(dtype_backend='pyarrow')
```

```python
sales_data.head(2)
```

```python
sales_data.dtypes
```

```python
#pd.set_option("display.float_format", "${:, .2f}".format)
sales_data.pivot_table(values="sale_price", index="sales_region", columns="order_type", aggfunc="sum",)
```

```python
#pd.set_option("display.float_format", "${:, .2f}".format)
sales_data.pivot_table(values="sale_price", index="sales_region", columns="order_type", aggfunc="sum", margins=True, margins_name="Totals")
```

```python
sales_data.pivot_table(values="sale_price", index="product_category", columns="sales_region", aggfunc="max")
```
